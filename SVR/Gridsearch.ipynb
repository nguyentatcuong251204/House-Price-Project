```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import SVR
from sklearn.metrics import r2_score
import pandas as pd
from sklearn.metrics.pairwise import rbf_kernel,polynomial_kernel,linear_kernel,sigmoid_kernel
```

```python
df=pd.read_csv(r"C:\Users\NTCUONG\notebook\cnt-first-project\feature_selection\training1.csv")
df
```

```python

model=SVR()
param_grid={
    "kernel":['rbf','poly'],
    "C":[0.001, 0.005, 0.01, 0.1, 0.15],
    "epsilon":[0.1, 0.3 , 0.15, 0.01, 0.5]
}
grid_search=GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring="r2")
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.2,random_state=42)
grid_search.fit(X_train,y_train)
print(f'Best parameters: {grid_search.best_params_}')

# Dá»± Ä‘oÃ¡n trÃªn táº­p kiá»ƒm tra
y_pred = grid_search.predict(X_test)

# TÃ­nh toÃ¡n lá»—i MSE
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {r2}')
```
